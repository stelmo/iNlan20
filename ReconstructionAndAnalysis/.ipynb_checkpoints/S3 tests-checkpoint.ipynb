```python
fba_sol = model.optimize()
fba_sol.fluxes.to_json("fluxes\\fbafluxes.json")

# pfba_sol = cobra.flux_analysis.pfba(model)
# pfba_sol.fluxes.to_json("fluxes\\pfbafluxes.json")

# model.summary(fva=0.9)
model.summary()
```

```python
mu = fba_sol.fluxes["Biomass"]
bm_rxn = model.reactions.get_by_id('Biomass')
mu_cons = model.problem.Constraint(bm_rxn.flux_expression, lb = 0.90*mu, ub=mu)
model.add_cons_vars(mu_cons)
s = cobra.sampling.sample(model, 2000, method="achr")
s.to_json("fluxes\\fluxsamples.json")
```

```python
print(np.mean(s["EX_h2_e"]))
print(np.std(s["EX_h2_e"]))
```

```python
def ratio(formate, acetate, ethanol):
    bot = acetate + ethanol
    if bot < 0.0001:
        return np.nan
    rv = formate/bot
    return rv

s2 = s.copy()
s2['ratio'] = s2.apply(lambda row: ratio(row["EX_for_e"], row["EX_ac_e"],row["EX_etoh_e"]), axis=1)
sns.boxplot(s2["ratio"])
plt.show() 
```
